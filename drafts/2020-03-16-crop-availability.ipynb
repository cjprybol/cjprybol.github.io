```julia
import Pkg
pkgs =
[
"Dates",
"uCSV",
"DataFrames"
]
Pkg.add(pkgs)
for pkg in pkgs
    eval(Meta.parse("import $pkg"))
end
```

```julia
dir="$(homedir())/projects/$(Dates.today())-crop-availability"
if !isdir(dir)
    mkdir(dir)
end
cd(dir)
```

```julia
data, header = uCSV.read("$(homedir())/google-drive/Farming/2020/annuals.tsv", delim='\t', header=1, quotes='"', typedetectrows=100)
annuals = DataFrames.DataFrame(data, Symbol.(header))
```

```julia
availability_matrix = Bool.(Matrix(annuals[!, 2:13]))
```

```julia
# TODO improve by swapping these hueristics for weeks to maturity from seeding windows
```

```julia
for (column_index, month) in zip(4:4, string.(names(annuals)[5:5]))
    println(month)
    for row_index in 1:size(availability_matrix, 1)
        if availability_matrix[row_index, column_index - 2] && availability_matrix[row_index, column_index - 1]
            println(annuals[row_index, :Crop])
        end
    end
    println()
end

for (column_index, month) in zip(5:12, string.(names(annuals)[6:13]))
    println(month)
    for row_index in 1:size(availability_matrix, 1)
        if availability_matrix[row_index, column_index - 2] | availability_matrix[row_index, column_index - 1]
            println(annuals[row_index, :Crop])
        end
    end
    println()
end
```
