# Why
Microbes are the simplest lifeforms. They turn the raw chemical building blocks on earth into life-sustaining compounds that support all of the plants, animals, and ecosystems on earth. When leveraged they can be some of our greatest allies, and when underestimated, they can create plagues of death and disease. The more we understand them, the more we can harness them to build sustainable and regenerative relationships with the natural world around us.

# What/Company purpose Start here: define your company in a single declarative sentence.
Software and services to sequence, assemble, analyze, and understand microbial populations.

# How
Maximum-likelihood, iterative graph genome assembly and genome-wide relative phylogenies for classification and quantification with presentable and digestible results and analysis specific to the customer's needs at a competitive price.

# Problem: Describe the pain of your customer. How is this addressed today and what are the shortcomings to current solutions.
There are no unbiased methods for analyzing and understanding entire populations of microbes targetting the consumer/commercial marketplace outside of the medical industry. Current offerings for microbial analysis are targetted, and look only for specific organisms of interest (when evaluating species present) or at non-specific measures of population structure (microscopy or population metabolic activity which gives no species information).

# Solution: Explain your eureka moment. Why is your value prop unique and compelling? Why will it endure? And where does it go from here?
DNA sequencing provides the most accurate and comprehensive means of assessing and understanding microbial populations. I have developed algorithms and methods that work with existing DNA sequencing technology and provide results with significantly less bias than existing methodologies. The methods will endure as open source material in the public domain. The service will endure by providing excellent user satisfaction and general purpose applicability.

# Why now? The best companies almost always have a clear why now? Nature hates a vacuum—so why hasn’t your solution been built before now? What’s true about the world that nobody else figured out yet?
The technology to sequence microbial populations exists and is sufficiently accurately and affordable. The computing power to analyze the data is also sufficiently available and affordable. The time is now because I have developed the algorithms, methods, and intellectual property to apply these resources to an unaddressed domain.

# Vision If all goes well, what will you have built in five years?
A service for general purpose microbial sequencing, assembling, and analyzing focusing on uncompromising customer satisfaction, utility, and open source tools.

# Why did you pick this idea to work on? Do you have domain expertise in this area? How do you know people need what you're making?
After studying biochemistry and ecology I felt that genetics was our best tool for monitoring the biological diversity and associated resilience of the environment around us. After studying genetics in a medical context I felt that a better understanding of microbes was our best path forward to improving both environmental and human health and productivity. I have a decade of experience researching and developing in biochemistry/microbiology/ecology and a half decade of Ph.D. level experience developing the most comprehensive set of genome assembly and analysis software available today. I know that people need what I'm making because I needed what I have made. As long as humans and the environment face challenges affected their health, there will be a need to leverage and understand microbes.

# What do you understand about your business that other companies in it just don't get?
Existing solutions for genome assembly and analysis are based on "greedy" methods and heuristics. In many cases greedy methods and heuristics provide sufficiently accurate solutions to problems given resource constraints, however, in the case of understanding genetics and in research where knowledge builds upon itself in a compounding way, these greedy and heuristic-based approaches create a foundation for errors to compound upon themselves and pervasively influence our understanding of human and environmental health and disease. I feel that these tradeoffs have led to many of the underwhelming results of the promise of precision medicine in the 21st century and a healthcare system that takes advantage of patients and consumers genetic information. I think these issues will play out in environmental health research as well if they are not addressed by more resilient analytical methods.
